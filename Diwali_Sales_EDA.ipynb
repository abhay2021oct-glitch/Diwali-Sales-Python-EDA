# Exploratory Data Analysis

## Gender

### Age

```python
xy = sns.countplot(data = df, x = 'Age Group', hue = 'Gender')
for bars in xy.containers:
    ax.bar_label(bars)
    plt.show()
```

```python
# Total Amount vs Age Group
sales_age = df.groupby(['Age Group'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)

sns.barplot(x = 'Age Group',y= 'Amount' ,data = sales_age)
```

*From above graphs we can see that most of the buyers are of age group between 26-35 yrs female*

### State

```python
# Total number of orders from top 10 states

sales_state = df.groupby(['State'], as_index=False)['Orders'].sum().sort_values(by='Orders', ascending=False).head(10)

sns.set(rc={'figure.figsize':(15,5)})
sns.set_theme(style="whitegrid",palette="viridis")
sns.barplot(data = sales_state, x = 'State',y= 'Orders')
```

```python
# total amount/sales from top 10 states
sales_state = df.groupby(['State'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False).head(10)

sns.set(rc={'figure.figsize':(15,5)})
sns.set_theme(style="darkgrid",palette="RdBu")
sns.barplot(data = sales_state, x = 'State',y= 'Amount')
```

*From above graphs we can see that most of the orders & total sales/amount are from Uttar Pradesh, Maharashtra and Karnataka respectively*

```python
ax = sns.countplot(data = df, x = 'Marital_Status')
sns.set(rc={'figure.figsize':(7,5)})
sns.set_theme(style="darkgrid",palette="RdBu")
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_state = df.groupby(['Marital_Status', 'Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)
sns.set(rc={"figure.figsize":(5,5)})
sns.barplot(data= sales_state, x = "Marital_Status", y = "Amount", hue = "Gender")
```

*From above graphs we can see that most of the buyers are married (women) and they have high purchasing power*

### Occupation

```python
sns.set(rc={"figure.figsize":(20,5)})
ax = sns.countplot(data = df, x = "Occupation")
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_state = df.groupby(['Occupation'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)

sns.set(rc={'figure.figsize':(20,5)})
sns.set_theme(style="darkgrid",palette="viridis")
sns.barplot(data = sales_state, x = 'Occupation',y= 'Amount')
```

*From above graphs we can see that most of the buyers are working in IT, Healthcare and Aviation sector*

### Product Category

```python
ax = sns.countplot(data = df, x = 'Product_Category')
sns.set(rc={'figure.figsize':(20,5)})
sns.set_theme(style="darkgrid",palette="RdBu")
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_state = df.groupby(['Product_Category'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False).head(10)
sns.set(rc={"figure.figsize":(20,5)})
sns.set_theme(style="darkgrid",palette="coolwarm")
sns.barplot(data= sales_state, x = "Product_Category",y = "Amount")
```

*From above graphs we can see that most of the sold products are from Food, Clothing and Electronics category*

```python
sales_state = df.groupby(['Product_ID'], as_index=False)['Orders'].sum().sort_values(by='Orders', ascending=False).head(10)

sns.set(rc={'figure.figsize':(20,5)})
sns.set_theme(style="darkgrid",palette="viridis")
sns.barplot(data = sales_state, x = 'Product_ID',y= 'Orders')
```

```python
# top 10 most sold products (same thing as above)

fig1, ax1 = plt.subplots(figsize=(15,7))
df.groupby('Product_ID')['Orders'].sum().nlargest(10).sort_values(ascending=False).plot(kind='bar')
```

```python
orders = df.groupby('Product_Category')['Orders'].sum()
sales = df.groupby('Product_Category')['Amount'].sum()

insight_df = pd.concat([orders, sales], axis=1).reset_index()
insight_df.head()

```

*Some product categories generate a large number of orders but contribute less to overall revenue due to lower average selling prices.*

```python
order_sales_df = df.groupby('Product_Category').agg(
    Total_Orders=('Orders', 'sum'),
    Total_Sales=('Amount', 'sum')
).reset_index()

plt.figure(figsize=(8,5))
sns.scatterplot(
    data=order_sales_df,
    x='Total_Orders',
    y='Total_Sales')

plt.title('Relationship Between Orders and Sales')
plt.xlabel('Total Orders')
plt.ylabel('Total Sales Amount')
plt.show()
```

*The relationship between total orders and total sales is not perfectly linear, indicating that higher order volume does not always guarantee higher revenue.*

## Business Recommendations

Based on the exploratory data analysis of Diwali sales data, the following recommendations can help improve business performance:

Focus on High-Revenue Categories
Product categories that generate high sales with relatively fewer orders should be prioritized during festive campaigns, as they contribute significantly to overall revenue.

Upsell Low Ticket-Size Products
Categories with high order volume but low revenue indicate low-priced products. Introducing bundles, combo offers, or cross-selling can help increase the average order value.

Target Key Customer Segments
Age groups and customer segments contributing the highest sales should be targeted with personalized offers and festive discounts to maximize conversion.

State-Level Marketing Strategy
High-performing states can be targeted with premium product promotions, while underperforming states may benefit from localized offers and increased marketing visibility.

Optimize Product Mix During Festive Sales
A balanced mix of high-volume and high-value products should be maintained to ensure both customer engagement and revenue growth.

## Conclusion:

*The analysis shows that Diwali sales performance is driven more by product pricing and category mix than by order volume alone. Additionally, married women aged 26–35 years from Uttar Pradesh, Maharashtra, and Karnataka, working in IT, Healthcare, and Aviation sectors, are more likely to purchase products from Food, Clothing, and Electronics categories. These insights can help businesses design targeted festive marketing and product strategies.*
